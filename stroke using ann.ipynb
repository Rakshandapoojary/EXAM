```python
df['bmi'].fillna(df['bmi'].mean(),inplace=True)
df
```

```python
df=df.drop(columns=['id'],axis=1)
df
```

```python
df.info()
```

```python
cat_features = ["gender","ever_married","work_type","Residence_type","smoking_status"]
df = pd.get_dummies(df,columns=cat_features)
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(df[["age","hypertension","heart_disease","avg_glucose_level","bmi","stroke"]]) 
data_normalized = scaler.transform(df[["age","hypertension","heart_disease","avg_glucose_level","bmi","stroke"]]) 
data_normalized
```

```python
data_df_normalized=pd.DataFrame(data_normalized,columns=["age","hypertension","heart_disease","avg_glucose_level","bmi","stroke"])
data_df_normalized=data_df_normalized.join(df[df.columns.drop(["age","hypertension","heart_disease","avg_glucose_level","bmi","stroke"])])
data_df_normalized
```

```python
Y = data_df_normalized['stroke']
X = data_df_normalized[data_df_normalized.columns.drop('stroke')]
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.33, random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
train_score=model.score(X_train,Y_train)
test_score=model.score(X_test,Y_test)
print('Train Score (R-Squared): ',train_score)
print('Test Score (R-Squared)',test_score)
```

```python
from sklearn.neural_network import MLPRegressor
model = MLPRegressor(hidden_layer_sizes=(50))
model.fit(X_train,Y_train)
predictions_digit= model.predict(X_train) 
print(X_train.shape)
```
