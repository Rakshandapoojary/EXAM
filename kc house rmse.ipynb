```python
data.loc[:,'year']=data['date'].str[0.4]
data.loc[:,'month']=data['date'].str[4.8]
data=data.drop(columns=['date'])
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit(data[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'floors']])
data_normalized=scaler.transform(data[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'floors']])
data_normalized
```

```python
data_normalized=pd.DataFrame(data_normalized,columns=['price', 'bedrooms', 'bathrooms', 'sqft_living', 'floors'])
data_normalized=data_normalized.join(data[data.columns.drop(['price', 'bedrooms', 'bathrooms', 'sqft_living', 'floors'])])
data_normalized                                                            
```

```python
cat_features=["waterfront", "view", "condition", "grade", "year", "month", "zipcode"]
data=pd.get_dummies(data,columns=cat_features)
data.columns
```

```python
x=data.drop(columns=['price'])
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
train_score=model.score(x_train,y_train)
test_score=model.score(x_train,y_train)
print("train (Rsquared):",train_score)
print("test(rsquared):",test_score)
```

```python
from sklearn.metrics import mean_squared_error
train_prediction=model.predict(x_train)
train_RMSE=mean_squared_error(y_train,train_prediction)**0.5
test_prediction=model.predict(x_test)
test_RMSE=mean_squared_error(y_test,test_prediction)**0.5
print("train rmse:",train_RMSE)
print("test rmse:",test_RMSE)
```
